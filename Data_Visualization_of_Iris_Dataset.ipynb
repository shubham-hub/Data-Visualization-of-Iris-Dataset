```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
iris = pd.read_csv('/content/drive/MyDrive/EDA Projects /Exploratory Data Analysis(EDA) of Iris Species Dataset/Iris.csv')
```

```python
iris.head()
```

```python
#print scatter plot
#for sepal length and petal length
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris)
plt.show()
```

```python
#map on basis of species for sepal length and petal length
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species')
```

```python
#adding hue to petal length to help differentiate 
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='PetalLengthCm')
plt.show()
```

```python
#adding hue to sepal length to help differentiate 
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='SepalLengthCm')
plt.show()
```

```python
#for sepal width and petal width
sns.scatterplot(x='SepalWidthCm',y='PetalWidthCm',data=iris,hue='Species')
plt.show()
```
